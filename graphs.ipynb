```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
values=[0.6,1.2,1.8,2.4,6,12]
file="/home/nassosliak/Desktop/avlsi/Breast_Cancer/cadence/l.csv"
data=pd.read_csv(file)
vin=data.iloc[:,0]
plt.figure()
plt.grid()
plt.xlabel("Vin(V)")
plt.ylabel("Iout(A)")
plt.title("Parametric Analysis for L")
j=0
for v in values:
    i=data.iloc[:,j+1]
    j+=1
    plt.plot(vin,i,label=str(v)+"u")
plt.legend()
```
